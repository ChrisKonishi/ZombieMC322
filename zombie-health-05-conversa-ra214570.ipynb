## Conversa entre Zumbis

Considere que zumbis podem conversar entre si através de um método especial `pergunta`, este método recebe uma pergunta (`String`) e retorna uma resposta (`String`). Podem ser feitas perguntas como a idade, nome e estado. Por exemplo, o método pode receber a pergunta:
~~~
"Qual a sua idade?"
~~~

e então retorna a idade.

Implemente este método nas três classes (`GarotoZumbi`, `Monstro` e `Perfeito`).

```java
public class GarotoZumbi
{
    protected int idade;
    protected String estado;
    protected String nome;
    
    public GarotoZumbi(int idade, String estado, String nome) {
        this.idade = idade;
        this.estado = estado;
        this.nome = nome;
    }
    
    public void mostra() {
        // cabeleira
        if (idade >= 2)
            System.out.println("  *");
        
        // corpo com olhos
        if (estado.equalsIgnoreCase("acordado"))
            System.out.println(" o*o");
        else
            System.out.println(" -*-");
        
        // barba
        if (idade >= 3)
            System.out.println("*****");
        
        System.out.println(nome);
        
        System.out.println();
    }
    
    public void cresce() {
        if (idade < 3)
            idade++;
        mostra();
    }
    
    public void acorda() {
        estado = "acordado";
        mostra();
    }
    
    public void dorme() {
        estado = "dormindo";
        mostra();
    }
    
    public String pergunta(String p) {
        String resposta = "";
        
        if (p.equalsIgnoreCase("Qual a sua idade?"))
            resposta = Integer.toString(idade);
        
        else if (p.equalsIgnoreCase("Qual o seu nome?"))
            resposta = nome;
        
        else if (p.equalsIgnoreCase("Qual o seu estado?"))
            resposta = estado;
        
        return resposta;
    }
}


public class Monstro extends GarotoZumbi{
    
    public Monstro(int idade, String estado, String nome){
        super(idade, estado, nome);
    }
    
    public void cresce_monstro(){
        if (idade <= 2)
            cresce();
        
        else if (idade < 5){
            idade++;
            mostra_monstro();
        }
    }
    
    public void mostra_monstro(){
        if (idade <= 3){
            mostra();
        }
        
        else {
            //cabelo
            System.out.println("  *");
            
            //olhos
            if (estado.equalsIgnoreCase("acordado"))
                System.out.println(" o*o");
            
            else 
                System.out.println(" -*-");
            
            //barba
            System.out.println("*****");
            
            if (idade >= 4)
                System.out.println("#####");
            
            if (idade >= 5)
                System.out.println("/   \\"); 
                               
            System.out.println(nome);
        
            System.out.println();
        }
    }
    
        public String pergunta(String p) {
        String resposta = "";
        
        if (p.equalsIgnoreCase("Qual a sua idade?"))
            resposta = Integer.toString(idade);
        
        else if (p.equalsIgnoreCase("Qual o seu nome?"))
            resposta = nome;
        
        else if (p.equalsIgnoreCase("Qual o seu estado?"))
            resposta = estado;
        
        return resposta;
    }
    
}


public class Perfeito extends GarotoZumbi{

    
    public Perfeito(int idade, String estado, String nome){
        super(idade, estado, nome);
    }
    
    public void mudaEstado(){
        
        if (estado.equalsIgnoreCase("normal"))
            estado = "milionario";
        
        else if (estado.equalsIgnoreCase("milionario"))
            estado = "doido";
        
        else
            estado = "normal";
        
        mostraPerfeito();
        
    }
    
    public void mostraPerfeito(){
        System.out.println("+-----+"); //cabelo
        
        if (idade >= 2)
            System.out.println("|     |"); //Testa
        
        //Agora os olhos
        if (estado.equalsIgnoreCase("normal"))
            System.out.println("| o-o |");
            
        else if (estado.equalsIgnoreCase("milionario"))
            System.out.println("| $-$ |");
            
        else
            System.out.println("| @-@ |");
        
        //Agora a boca
        if (estado.equalsIgnoreCase("doido"))
            System.out.println("|  ~  |");
        
        else
            System.out.println("| ___ |");
        
        //entre boca e queixo
        if (idade >= 3)
            System.out.println("|     |");
        
        //Agora o queixo!!
        if (idade <= 2)
            System.out.println(" \\___/ ");
        
        else
            System.out.println("\\_____/");
        
        
        System.out.println(nome);
    }
    
    void cresce_sem_print(){
        if (idade < 3)
            idade++;
    }
    
        public String pergunta(String p) {
        String resposta = "";
        
        if (p.equalsIgnoreCase("Qual a sua idade?"))
            resposta = Integer.toString(idade);
        
        else if (p.equalsIgnoreCase("Qual o seu nome?"))
            resposta = nome;
        
        else if (p.equalsIgnoreCase("Qual o seu estado?"))
            resposta = estado;
        
        return resposta;
    }
}


Perfeito garotoP = new Perfeito(1, "milionario", "Joseph");
Monstro garotoM = new Monstro(1, "acordado", "Asdrubal");
GarotoZumbi garoto = new GarotoZumbi(2, "dormindo", "Umaru");

System.out.println(garotoP.pergunta("Qual o seu nome?"));
System.out.println(garotoP.pergunta("Qual a sua idade?"));
System.out.println(garotoP.pergunta("Qual o seu estado?"));

System.out.println(garotoM.pergunta("Qual o seu nome?"));
System.out.println(garotoM.pergunta("Qual a sua idade?"));
System.out.println(garotoM.pergunta("Qual o seu estado?"));

System.out.println(garoto.pergunta("Qual o seu nome?"));
System.out.println(garoto.pergunta("Qual a sua idade?"));
System.out.println(garoto.pergunta("Qual o seu estado?"));
```

## Engajando zumbis na conversa

Escreva um método que permita *engajar* dois objetos zumbi em uma conversa. Após engajar dois zumbis, um deles deve ser capaz de contactar o outro para fazer perguntas.

Construa uma classe zumbi entrevistador que faça perguntas a outros zumbis para selecionar o mais velho entre eles. Se dois ou mais zumbis têm a mesma idade é dado preferência ao primeiro.

Escreva um programa em que um zumbi entrevistador entreviste três zumbis candidatos. Ele deve imprimir uma mensagem indicando o nome do mais velho entre os zumbis. Você deve *engajar*  cada um dos candidatos com o entrevistador, que informará o resultado quando o método `resultado` for acionado.

```java
public class Entrevistador{
    private String nome; //Dados do mais velho
    private int idade;
    
    public Entrevistador(){
        nome = "";
        idade = 0;
    }
    
    public void engajar(GarotoZumbi g){
        String j;
        int i;
        
        
        j = g.pergunta("Qual a sua idade?");
        i = Integer.parseInt(j);
        
        if (i > idade){
            idade = i;
            nome = g.pergunta("Qual o seu nome?");
        }
    }
    
    public void engajar(Monstro g){
        String j;
        int i;
        
        
        j = g.pergunta("Qual a sua idade?");
        i = Integer.parseInt(j);
        
        if (i > idade){
            idade = i;
            nome = g.pergunta("Qual o seu nome?");
        }
    }
    
    public void engajar(Perfeito g){
        String j;
        int i;
        
        
        j = g.pergunta("Qual a sua idade?");
        i = Integer.parseInt(j);
        
        if (i > idade){
            idade = i;
            nome = g.pergunta("Qual o seu nome?");
        }
    }
    
    
    
    public void resultado(){
        System.out.println("O zumbi mais velho é: " + nome);
    }
}

Perfeito garotoP = new Perfeito(3, "milionario", "Joseph");
Monstro garotoM = new Monstro(1, "acordado", "Asdrubal");
GarotoZumbi garoto = new GarotoZumbi(2, "dormindo", "Umaru");
Entrevistador e = new Entrevistador();

e.engajar(garoto);
e.engajar(garotoP);
e.engajar(garotoM);
e.resultado();
```
